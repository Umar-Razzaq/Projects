#### Create a contigency table for Attrition and JobSatisfaction columns. and comment on the results in a markdown

Most of the values for Attrition are "NO". Moreover the job satisfaction increases for people with Attrition=No.

#### Perform a chi square test for Attrition and JobSatisfaction columns. And comment on the results in a markdown

```python
from scipy import stats
c, p, dof, expected = stats.chi2_contingency(pd.crosstab(hr['Attrition'],hr['JobSatisfaction']))
print('chi_square:',c)
print('p value:',p)
print('degree of freedom:',dof)
```

As p value is less tha 5% hence there is sufficient evidence to conclude that the observed distribution is not the same as the expected distribution.
