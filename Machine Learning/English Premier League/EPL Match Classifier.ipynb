```python
def standingInLastSeason(dfs, standings):
    count=0
    for df in dfs:
        hs = np.zeros(len(df))
        aws = np.zeros(len(df))

        for i, match in df.iterrows():
            homeTeam=match.HomeTeam
            awayTeam=match.AwayTeam
            
            homeStanding=standings.loc[standings['teams'] == homeTeam].iloc[:, count]
            if len(homeStanding)==0:
                hs[i]=21
            else:
                hs[i]=int(homeStanding)
            print(homeTeam, homeStanding, count, i)
            
            awayStanding=standings.loc[standings['teams'] == awayTeam].iloc[:, count]
            if len(awayStanding)==0:
                aws[i]=21
            else:
                aws[i]=int(awayStanding)
#             print(awayTeam, awayStanding)

        df['Home Team Standing In Last Season']=hs
        df['Away Team Standing In Last Season']=aws
        count+=1
```

```python
def goals_by_each_team(df):
    hg = np.zeros(len(df))
    ag = np.zeros(len(df))
    hgc= np.zeros(len(df))
    agc= np.zeros(len(df))
    
    for i, match in df.iterrows():
        homeTeam=match.HomeTeam
        awayTeam=match.AwayTeam
        newdf=df.iloc[:i].copy()
        
        homeGoalsTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].FTHG.sum()
        homeGoalsTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].FTAG.sum()
        totalGoalsByHomeTeam=homeGoalsTillMatchAtAway+homeGoalsTillMatchAtHome
        hg[i]=totalGoalsByHomeTeam
        
        homeGoalsConcededTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].FTAG.sum()
        homeGoalsConcededTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].FTHG.sum()
        totalGoalsConcededByHomeTeam=homeGoalsConcededTillMatchAtAway+homeGoalsConcededTillMatchAtHome
        hgc[i]=totalGoalsConcededByHomeTeam
        
        awayGoalsTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].FTHG.sum()
        awayGoalsTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].FTAG.sum()
        totalGoalsByAwayTeam=awayGoalsTillMatchAtAway+awayGoalsTillMatchAtHome
        ag[i]=totalGoalsByAwayTeam
        
        awayGoalsConcededTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].FTAG.sum()
        awayGoalsConcededTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].FTHG.sum()
        totalGoalsConcededByAwayTeam=awayGoalsConcededTillMatchAtAway+awayGoalsConcededTillMatchAtHome
        agc[i]=totalGoalsConcededByAwayTeam
        
        
             
    df['Total Goals By Home Team In Season']=hg
    df['Total Goals By Away Team In Season']=ag
    
    df['Total Goals Conceded By Home Team In Season']=hgc
    df['Total Goals Conceded By Away Team In Season']=agc
```

```python
def points_of_each_team(df):
    pointsOfHome=np.zeros(len(df))
    pointsOfAway=np.zeros(len(df))
    
    for i, match in df.iterrows():
        homeTeam=match.HomeTeam
        awayTeam=match.AwayTeam
        newdf=df.iloc[:i].copy()

        goalsathome=newdf[newdf['HomeTeam']==homeTeam].copy()
        goalsathome.reset_index(drop=True, inplace=True)
        temp=[]
        for j in range(len(goalsathome)):
            if goalsathome.iloc[j].FTHG > goalsathome.iloc[j].FTAG:
                temp.append(3)
            if goalsathome.iloc[j].FTHG < goalsathome.iloc[j].FTAG:
                temp.append(0)
            if goalsathome.iloc[j].FTHG == goalsathome.iloc[j].FTAG:
                temp.append(1)
        pointAtHomeByHome=sum(temp)
        
        goalsataway=newdf[newdf['AwayTeam']==homeTeam].copy()
        goalsataway.reset_index(drop=True, inplace=True)
        temp=[]
        for j in range(len(goalsataway)):
            if goalsataway.iloc[j].FTHG > goalsataway.iloc[j].FTAG:
                temp.append(0)
            if goalsataway.iloc[j].FTHG < goalsataway.iloc[j].FTAG:
                temp.append(3)
            if goalsataway.iloc[j].FTHG == goalsataway.iloc[j].FTAG:
                temp.append(1)
        pointAtAwayByHome=sum(temp)
        
        pointsHome=pointAtAwayByHome+pointAtHomeByHome
        pointsOfHome[i]=pointsHome
        
        goalsathome=newdf[newdf['HomeTeam']==awayTeam].copy()
        goalsathome.reset_index(drop=True, inplace=True)
        temp=[]
        for j in range(len(goalsathome)):
            if goalsathome.iloc[j].FTHG > goalsathome.iloc[j].FTAG:
                temp.append(3)
            if goalsathome.iloc[j].FTHG < goalsathome.iloc[j].FTAG:
                temp.append(0)
            if goalsathome.iloc[j].FTHG == goalsathome.iloc[j].FTAG:
                temp.append(1)
        pointAtHomeByAway=sum(temp)
        
        goalsataway=newdf[newdf['AwayTeam']==awayTeam].copy()
        goalsataway.reset_index(drop=True, inplace=True)
        temp=[]
        for j in range(len(goalsataway)):
            if goalsataway.iloc[j].FTHG > goalsataway.iloc[j].FTAG:
                temp.append(0)
            if goalsataway.iloc[j].FTHG < goalsataway.iloc[j].FTAG:
                temp.append(3)
            if goalsataway.iloc[j].FTHG == goalsataway.iloc[j].FTAG:
                temp.append(1)
        pointAtAwayByAway=sum(temp)
        
        pointsAway=pointAtAwayByAway+pointAtHomeByAway
        pointsOfAway[i]=pointsAway

    df['Points Of Home Team In Season']=pointsOfHome
    df['Points Of Away Team In Season']=pointsOfAway
```

```python
def points_of_each_team_in_last5(df):
    pointsOfHome=np.zeros(len(df))
    pointsOfAway=np.zeros(len(df))
    
    for i, match in df.iterrows():
        homeTeam=match.HomeTeam
        awayTeam=match.AwayTeam
        newdf=df.iloc[:i+1].copy()
        
        goalsofhome=newdf[(newdf.HomeTeam == homeTeam) | (newdf.AwayTeam == homeTeam)].copy()
        goalsofhome.reset_index(drop=True, inplace=True)
        temp=[]
        
        for j in range(len(goalsofhome)):
            if goalsofhome.iloc[j].HomeTeam==homeTeam:
                temp.append(goalsofhome.iloc[j]['Points Of Home Team In Season'])
            if goalsofhome.iloc[j].AwayTeam==homeTeam:
                temp.append(goalsofhome.iloc[j]['Points Of Away Team In Season'])
        temp=temp[-6:]
        temp=np.diff(temp)
        pointsOfHome[i]=sum(temp)
        
        goalsofaway=newdf[(newdf['HomeTeam']==awayTeam)|(newdf['AwayTeam']==awayTeam)].copy()
        goalsofaway.reset_index(drop=True, inplace=True)
        temp=[]
        for j in range(len(goalsofaway)):
            if goalsofaway.iloc[j].HomeTeam==awayTeam:
                temp.append(goalsofaway.iloc[j]['Points Of Home Team In Season'])
            if goalsofaway.iloc[j].AwayTeam==awayTeam:
                temp.append(goalsofaway.iloc[j]['Points Of Away Team In Season'])
        temp=temp[-6:]
        temp=np.diff(temp)
        pointsOfAway[i]=sum(temp)
        
    df['Points Of Home Team In Last 5 Games Of Season']=pointsOfHome
    df['Points Of Away Team In Last 5 Games Of Season']=pointsOfAway
```

```python
def home_wld_yvar(df):
    var=[]
    for i, match in df.iterrows():
        if match.FTHG>match.FTAG:
            var.append('Hwin')
        if match.FTHG<match.FTAG:
            var.append('Hloose')
        if match.FTHG==match.FTAG:
            var.append('Hdraw')
    df['Home Result']=var
```

```python
def various(df):
    hsh = np.zeros(len(df))
    hshcon = np.zeros(len(df))
    ash = np.zeros(len(df))
    ashcon = np.zeros(len(df))
    
    hsht = np.zeros(len(df))
    hshtcon = np.zeros(len(df))
    asht = np.zeros(len(df))
    ashtcon = np.zeros(len(df))
    
    hhw = np.zeros(len(df))
    hhwcon = np.zeros(len(df))
    ahw = np.zeros(len(df))
    ahwcon = np.zeros(len(df))
    
    hc=np.zeros(len(df))
    hccon=np.zeros(len(df))
    ac=np.zeros(len(df))
    accon=np.zeros(len(df))
    
    hoff=np.zeros(len(df))
    aoff=np.zeros(len(df))
    
    hred=np.zeros(len(df))
    ared=np.zeros(len(df))
    
    
    for i, match in df.iterrows():
        homeTeam=match.HomeTeam
        awayTeam=match.AwayTeam
        newdf=df.iloc[:i].copy()
                
        
        homeRedTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].HR.sum()
        homeRedTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].AR.sum()
        totalRedByHomeTeam=homeRedTillMatchAtAway+homeRedTillMatchAtHome
        hred[i]=totalRedByHomeTeam
        
        awayRedTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].HR.sum()
        awayRedTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].AR.sum()
        totalRedByAwayTeam=awayRedTillMatchAtAway+awayRedTillMatchAtHome
        ared[i]=totalRedByAwayTeam
        
        
        
        
        homeShotsTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].HS.sum()
        homeShotsTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].AS.sum()
        totalShotsByHomeTeam=homeShotsTillMatchAtAway+homeShotsTillMatchAtHome
        hsh[i]=totalShotsByHomeTeam
        
        homeShotsConcededTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].AS.sum()
        homeShotsConcededTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].HS.sum()
        totalShotsConcededByHomeTeam=homeShotsConcededTillMatchAtAway+homeShotsConcededTillMatchAtHome
        hshcon[i]=totalShotsConcededByHomeTeam
        
        awayShotsTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].HS.sum()
        awayShotsTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].AS.sum()
        totalShotsByAwayTeam=awayShotsTillMatchAtAway+awayShotsTillMatchAtHome
        ash[i]=totalShotsByAwayTeam
        
        awayShotsConcededTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].AS.sum()
        awayShotsConcededTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].HS.sum()
        totalShotsConcededByAwayTeam=awayShotsConcededTillMatchAtAway+awayShotsConcededTillMatchAtHome
        ashcon[i]=totalShotsConcededByAwayTeam
        
        
        
        
        homeShotsOnTTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].HST.sum()
        homeShotsOnTTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].AST.sum()
        totalShotsOnTByHomeTeam=homeShotsOnTTillMatchAtAway+homeShotsOnTTillMatchAtHome
        hsht[i]=totalShotsOnTByHomeTeam
        
        homeShotsOnTConcededTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].AST.sum()
        homeShotsOnTConcededTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].HST.sum()
        totalShotsOnTConcededByHomeTeam=homeShotsOnTConcededTillMatchAtAway+homeShotsOnTConcededTillMatchAtHome
        hshtcon[i]=totalShotsOnTConcededByHomeTeam
        
        awayShotsOnTTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].HST.sum()
        awayShotsOnTTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].AST.sum()
        totalShotsOnTByAwayTeam=awayShotsOnTTillMatchAtAway+awayShotsOnTTillMatchAtHome
        asht[i]=totalShotsOnTByAwayTeam
        
        awayShotsOnTConcededTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].AST.sum()
        awayShotsOnTConcededTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].HST.sum()
        totalShotsOnTConcededByAwayTeam=awayShotsOnTConcededTillMatchAtAway+awayShotsOnTConcededTillMatchAtHome
        ashtcon[i]=totalShotsOnTConcededByAwayTeam
        
        
        
        

        homeCornersTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].HC.sum()
        homeCornersTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].AC.sum()
        totalCornersByHomeTeam=homeCornersTillMatchAtAway+homeCornersTillMatchAtHome
        hc[i]=totalCornersByHomeTeam
        
        homeCornersConcededTillMatchAtHome=newdf[newdf.HomeTeam == homeTeam].AC.sum()
        homeCornersConcededTillMatchAtAway=newdf[newdf.AwayTeam == homeTeam].HC.sum()
        totalCornersConcededByHomeTeam=homeCornersConcededTillMatchAtAway+homeCornersConcededTillMatchAtHome
        hccon[i]=totalCornersConcededByHomeTeam
        
        awayCornersTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].HC.sum()
        awayCornersTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].AC.sum()
        totalCornersByAwayTeam=awayCornersTillMatchAtAway+awayCornersTillMatchAtHome
        ac[i]=totalCornersByAwayTeam
        
        awayCornersConcededTillMatchAtHome=newdf[newdf.HomeTeam == awayTeam].AC.sum()
        awayCornersConcededTillMatchAtAway=newdf[newdf.AwayTeam == awayTeam].HC.sum()
        totalCornersConcededByAwayTeam=awayCornersConcededTillMatchAtAway+awayCornersConcededTillMatchAtHome
        accon[i]=totalCornersConcededByAwayTeam   
        
             
    df['Total Shots By Home Team In Season']=hsh
    df['Total Shots By Away Team In Season']=ash
    
    df['Total Shots Conceded By Home Team In Season']=hshcon
    df['Total Shots Conceded By Away Team In Season']=ashcon
    
    
    
    df['Total Shots On Target By Home Team In Season']=hsht
    df['Total Shots On Target By Away Team In Season']=asht
    
    df['Total Shots On Target Conceded By Home Team In Season']=hshtcon
    df['Total Shots On Target Conceded By Away Team In Season']=ashtcon
    
    
    

    df['Total Corners By Home Team In Season']=hc
    df['Total Corners By Away Team In Season']=ac
    
    df['Total Corners Conceded By Home Team In Season']=hccon
    df['Total Corners Conceded By Away Team In Season']=accon
    
    

    df['Total Red Cards By Home Team In Season']=hred
    df['Total Red Cards By Away Team In Season']=ared
```

```python
def allinone(df):
    goals_by_each_team(df) #goals by both away and home team in that season till that match
    points_of_each_team(df) #points of both away and home team in that season till that match
    points_of_each_team_in_last5(df) #points of both away and home team in the 5 preceding games of the season (form of teams)
    home_wld_yvar(df) #various match stats of each team in that season uptil that match
    various(df)
```
